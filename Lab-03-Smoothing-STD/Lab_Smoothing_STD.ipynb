```python
io.imshow(noisy_img)
```

```python
'''
(1) Your own median filter algorithm
'''
def generate_median_filter(img, filter_size):
    
    '''
    @parameters:
        [image]: image with any shape
        [filter_size] : an integer which will result to (filter_size, filter_size) shaped matrix
    
    @returns:
        the filtered image
    '''
    img = np.pad(img, 1)
    output_img = np.zeros_like(img)
    
    dx = filter_size
    dy = filter_size
    
    for i in range(0, img.shape[0] - dx + 1, 1):
        for j in range(0, img.shape[1] - dy + 1, 1):
            output_img[i, j] = sorted(img[i:i + dx, j: j + dy].flatten())[4]

    return output_img
    
med_normal_img = generate_median_filter(noisy_img, 3)
io.imshow(med_normal_img)
```

```python
'''
(2) skimage median filter
'''

def apply_median_filter(image):
    return median(image)

med_image = apply_median_filter(noisy_img)
io.imshow(med_image)
```

```python
'''
(3) Gaussion Filters
'''
from skimage.filters import gaussian
def apply_Gaussion_filter(image, sigma):
    return gaussian(image, sigma)

gaussian = apply_Gaussion_filter(noisy_img, 2)
io.imshow(gaussian)
```
