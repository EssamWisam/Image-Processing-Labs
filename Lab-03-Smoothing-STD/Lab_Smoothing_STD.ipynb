```python
'''
(1) Your own median filter algorithm
'''
def generate_median_filter(image, filter_size):
    
    '''
    @parameters:
        [image]: image with any shape
        [filter_size] : an integer which will result to (filter_size, filter_size) shaped matrix
    
    @returns:
        the filtered image
    '''
    temp_data = []
    idx = filter_size // 2
    result = []
    result = np.zeros((len(img),len(img[0])))
    for i in range(len(img)):

        for j in range(len(img[0])):

            for z in range(filter_size):
                if i + z - idx < 0 or i + z - idx > len(img) - 1:
                    for c in range(filter_size):
                        temp_data.append(0)
                else:
                    if j + z - idx < 0 or j + idx > len(img[0]) - 1:
                        temp_data.append(0)
                    else:
                        for k in range(filter_size):
                            temp_data.append(img[i + z - idx][j + k - idx])

            temp_data.sort()
            result[i][j] = temp_data[len(temp_data) // 2]
            temp_data = []
    return result
    
med_normal_img = generate_median_filter(img, 3)
io.imshow(med_normal_img)
```

```python
'''
(2) skimage median filter
'''

def apply_median_filter(image):
    return median(image)

med_image = apply_median_filter(img)
io.imshow(med_image)
```

```python
'''
(3) Gaussion Filters
'''
from skimage.filters import gaussian
def apply_Gaussion_filter(image):
    return gaussian(image)

gaussian = apply_Gaussion_filter(img)
io.imshow(gaussian)
```
